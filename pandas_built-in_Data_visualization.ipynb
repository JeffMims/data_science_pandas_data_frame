```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

# Styles

```python
sns.countplot(x='sex',data=df_tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=df_tips)
```

```python
sns.countplot(x='sex',data=df_tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=df_tips)
sns.despine(left=True)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=df_tips)
#sns.despine()
```

# Scale and Context

```python
plt.figure(figsize=(12,3))
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=df_tips, palette='coolwarm')
```
