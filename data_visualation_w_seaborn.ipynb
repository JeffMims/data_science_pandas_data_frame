# Distribution Plots

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#!pip install seaborn
```

# Data
seaborn comes with built-in data sets

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

```python
df_tips.info()
```

```python
df_tips.describe
```

# Distplot

```python
sns.distplot(df_tips['total_bill'])
# Safe to ignore the warning
```

```python
sns.distplot(df_tips['total_bill'], kde=False, bins=30)
# Safe to ignore the warning
```

# JointPlot

```python
sns.jointplot(x='total_bill', y='tip', data=df_tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df_tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df_tips, kind='reg')
```

# PairPlot

```python
sns.pairplot(df_tips,  hue='sex', palette='coolwarm')
```

# Rugplot

```python
sns.rugplot(df_tips['total_bill'])
```

# Barplot and countpot

```python
sns.barplot(x='sex', y='total_bill', data=df_tips)
```

```python
sns.barplot(x='sex', y='tip', data=df_tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=df_tips, estimator=np.std)
```

# CountPlot

```python
sns.countplot(x='sex', data=df_tips)
```

# BoxPlot and ViolinPlot

```python
sns.boxplot(x='day',y='total_bill', data=df_tips,palette='rainbow')
```

```python
sns.boxplot(data=df_tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill', data=df_tips,palette='coolwarm', hue='smoker')
```

# ViolinPlot

```python
sns.violinplot(x='day',y='total_bill', data=df_tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill', data=df_tips,palette='rainbow', hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill', data=df_tips,palette='Set1', hue='sex', split=True)
```

# StripPlot

```python
sns.stripplot(x='day',y='total_bill', data=df_tips)
```

```python
sns.stripplot(x='day',y='total_bill', data=df_tips, jitter=True, hue='sex', palette='Set2',split=True)# you can use jitter=true
```

# Swarmplot

```python
sns.swarmplot(x='day',y='total_bill', data=df_tips)
```

```python
sns.swarmplot(x='day',y='total_bill', data=df_tips, hue='sex', palette='Set2')
```

```python
sns.swarmplot(x='day',y='total_bill', data=df_tips, hue='sex', palette='Set2',split=True)
```

# Combining Categorical Plots

```python
sns.violinplot(x='tip', y='day', data=df_tips, palette='rainbow')
sns.swarmplot(x='tip',y='day', data=df_tips, color='black', size=3)
```

# FactorPlot

```python
sns.factorplot(x='sex',y='total_bill', data=df_tips, kind='bar')
```

```python
sns.factorplot(x='sex',y='total_bill', data=df_tips, kind='bar')
```

# HetMap

```python
sns.heatmap(df_tips.corr(),cmap='coolwarm')
```

```python
sns.heatmap(df_tips.corr(),cmap='coolwarm',annot=True)
```

```python
flights = df_flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights, cmap='magma',  linecolor='white', linewidths=1, annot=False)
```
